```python
from sklearn.datasets import make_regression
import numpy as np
```

```python
X,y = make_regression(n_samples=4,n_features=1,n_informative=1,n_targets=1,noise=80,random_state=13)
```

```python
X
```

```python
y
```

```python
import matplotlib.pyplot as plt
plt.scatter(X,y)
```

```python
# Lets apply OLS
from sklearn.linear_model import LinearRegression
```

```python
reg = LinearRegression()
reg.fit(X,y)
```

```python
reg.coef_
```

```python
reg.intercept_
```

```python
plt.scatter(X,y)
plt.plot(X,reg.predict(X),color = "red")
```

# Applying Gradient Descent assuming slope is constant m = 78.35
# and assume starting value of intercept b = 0

```python
y_pred = ((78.35*X)+0).reshape(4)
```

```python
y_pred
```

```python
plt.scatter(X,y)
plt.plot(X,reg.predict(X),color="Blue",label="OLS")
plt.plot(X,y_pred,color="Green",label='b = 0')
plt.legend()
plt.show()
```

```python
#iteration 1
m = 78.35
b = 0

loss_slope = -2 * np.sum(y - m*X.ravel() - b)
loss_slope
```

```python
# lets take learning rate = 0.1
lr = 0.1

step_size = loss_slope*lr
step_size
```

```python
#calculating the new intercept 

b = b - step_size
b
```

```python
y_pred1 = ((78.35 * X) + b).reshape(4)

plt.scatter(X,y)
plt.plot(X,reg.predict(X),color="Blue",label = "OLS")
plt.plot(X,y_pred1,color="Black",label="b = {}".format(b))
plt.plot(X,y_pred,color="Green",label='b = 0')
plt.legend()
plt.show()
```

```python
#iteration 2
loss_slope = -2 * np.sum(y - m*X.ravel() - b)
loss_slope
```

```python
step_size = loss_slope*lr
step_size
```

```python
b = b - step_size
b
```

```python
y_pred1 = ((78.35 * X) + b).reshape(4)

plt.scatter(X,y)
plt.plot(X,reg.predict(X),color="Blue",label = "OLS")
plt.plot(X,y_pred1,color="Black",label="b = {}".format(b))
plt.plot(X,y_pred,color="Green",label='b = 0')
plt.legend()
plt.show()
```

```python
#iteration 3
loss_slope = -2 * np.sum(y - m*X.ravel() - b)
loss_slope
```

```python
step_size = loss_slope*lr
step_size
```

```python
b = b - step_size
b
```

```python
y_pred1 = ((78.35 * X) + b).reshape(4)

plt.scatter(X,y)
plt.plot(X,reg.predict(X),color="Blue",label = "OLS")
plt.plot(X,y_pred1,color="Black",label="b = {}".format(b))
plt.plot(X,y_pred,color="Green",label='b = 0')
plt.legend()
plt.show()
```

```python
# Through loop

b= 100
m = 78.35
lr = 0.01

epochs = 10

for i in range(epochs):
    loss_slope = -2 * np.sum(y - m*X.ravel() - b)
    b = b - (lr* loss_slope)
    y_pred = m*X + b
    plt.plot(X,y_pred)
plt.scatter(X,y)
```
